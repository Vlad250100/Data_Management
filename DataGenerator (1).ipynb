```python
import pandas as pd
import datetime
from datetime import timedelta
import datetime, calendar
pd.options.mode.chained_assignment = None

def add_months(sourcedate, months):
    month = sourcedate.month - 1 + months
    year = sourcedate.year + month // 12
    month = month % 12 + 1
    day = min(sourcedate.day, calendar.monthrange(year, month)[1])
    return datetime.date(year, month, day)


transactions = pd.DataFrame(columns = ['TRANSACTIONID','amount','datetime'])
subsriptions = pd.DataFrame(columns = ['ENDDATE','STARTDATE','TRANSACTIONID','USERID'])
tran_id = 1
amount = [i*100 for i in range(1, 11)]
for year in range(0,3):
 for month in range(1,13):
    for user in range(1,41):
      if random.randint(0, 100) > 60:
        today = datetime.datetime(2021+year, month, random.randint(1,28), random.randint(1,23))
        tran = amount[random.randint(0, 9)]
        user_sub = subsriptions[(subsriptions.USERID == user) & (subsriptions.ENDDATE >= today.date())]
        transactions.loc[len(transactions.index)] = [tran_id, tran, today]
        if len(user_sub) == 0:
          subsriptions.loc[len(subsriptions.index)] = [add_months(today, int(tran/100)), today.date(), tran_id, user ]
        else:
          new_date = user_sub['ENDDATE'].max()
          subsriptions.loc[len(subsriptions.index)] = [add_months(new_date , int(tran/100)), new_date + timedelta(days = 1), tran_id, user ]
        tran_id += 1
transactions_out = 'INSERT INTO Transaction (TRANSACTIONID, amount, datetime) VALUES '
subsriptions_out = 'INSERT INTO SUBSCRIPTION (ENDDATE, STARTDATE, TRANSACTIONID, USERID) VALUES '
for row in transactions.values:
  transactions_out += f"({row[0]}, {row[1]}, '{row[2]}'), "

for row in subsriptions.values:
  subsriptions_out += f"('{row[0]}', '{row[1]}', {row[2]}, {row[3]}), "
```
